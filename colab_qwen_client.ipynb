# اختبار واجهة Qwen-14B على Modal (Google Colab)

هذا الـ Notebook يوضّح كيف تنادي endpoint المنشور على modal.run بعد أن تَنشر التطبيق.
خطوات سريعة:
1. تأكد أنك نشرت الملف `modal_qwen_deploy.py` على Modal وحصلت على عنوان الـ endpoint (مثال: https://<YOUR-MODAL-APP-ID>.modal.run/).
2. افتح هذا الملف في Google Colab (File > Open notebook > GitHub) أو انسخ الرابط إلى Colab.
3. عدّل متغير `ENDPOINT` في الخلية التالية إلى عنوانك ثم شغّل الخلايا.


```python
# ضع هنا عنوان الendpoint الذي تحصل عليه بعد `modal deploy`
ENDPOINT = "https://<YOUR-MODAL-APP-ID>.modal.run/"  # استبدلها بعنوانك

import requests, json

def generate(prompt, model=None, max_new_tokens=128, timeout=120):
    payload = {
        'prompt': prompt,
        'max_new_tokens': max_new_tokens,
    }
    if model:
        payload['model'] = model
    resp = requests.post(ENDPOINT, json=payload, timeout=timeout)
    try:
        return resp.json()
    except Exception:
        return {'status_code': resp.status_code, 'text': resp.text}

print('جاهز — غيّر ENDPOINT ثم نفّذ الخلايا لاختبار الـ endpoint')
```

```python
# مثال اختبار سريع
prompt = "Write a short Arabic greeting and one-line description about AI."
res = generate(prompt, max_new_tokens=80)
print(json.dumps(res, ensure_ascii=False, indent=2))
```
